```java
Boolean a; 
Boolean b = false; 
Boolean c = true; 

System.out.println(a);
System.out.println(b);
System.out.println(c);
```

Javada sayılar int, double vb. türlere ayrılır. int türü tam sayı olan sayıları kapsar. Genişliği [-2,147,483,648, 2,147,483,647] büyüklüğündedir. Default olarak 0 değerine sahiptir. 

```java
int a;
int n1 = 10;
int n2 = 2147483647;
int n3 = 2147483647 + 1; //Maximum sayıdan sonra sayı negative'e dönmeye başlar. 
System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
```

Byte türü 8-bit'lik tam sayıları kapsayan türdür. Değeri -128 ile 127 arasında değişir. Default değeri 0'dır. Büyük arraylerde daha az hafıza kullanmak için kullanılır. integer türünden 4 kat daha küçüktür.


```java
byte a; 
byte n1 = 10; 
byte n2 = 127; 
byte n3 = (byte) 128; //128 maximum değeri aştığı için olmadığı için byte türüne çevirmeliyiz.  
byte n4 = (byte) (n1 + n1); //iki tane byte'ın toplamı byte türünde değil int türünde olacaktır. Bu değeri tekrar byte'a çevirmek zorundayız.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
```

Short türü 16-bit'lik tam sayıları kapsayan türdür. Maximum değeri 32.767'dir. Default değeri 0'dır. Short türü daha az hafıza kullanmak için kullanılabilir. integer türünden 2 kat daha küçüktür. 

```java
short a; 
short n1 = 10; 
short n2 = (short) (Math.pow(2, 15) - 1);
short n3 = (short) Math.pow(2, 15); // maximum değerden sonra negative'e dönmeye başlar.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
```

Float türü 32-bit'lik float halindeki sayıları kapsar. Alabileceği maximum değer 2^127'dir. Bu değerden sonra infinity değerini geri döndürür. Daha az hafıza kullanmak istiyorsak array'lerde double yerine float kullanmak daha mantıklıdır. Default değeri 0.0F'dir.

```java
float a; 
float n1 = 10; 
float n2 = (float)10.0; //nokta ile sayı tanımlandığında bu sayı double türüne girer. Bu sayıyı float'a çevirmeliyiz. 
float n3 = (float)Math.pow(2, 31);
float n4 = (float)Math.pow(2, 127);
float n5 = (float)Math.pow(2, 128);
float n6 = (float)100.0/(float)15.0;

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
System.out.println(n5);
System.out.println(n6);
```

Double türü ise 64-bit'lik float halinde olan sayıları kapsar. Alabileceği maximum değer 1.797,693,134,862,315,7 E+308 (2^1023) sayısıdır. Bu değerden sonra Infinity değerini geri döndürür. Default değeri 0.0d'dir. 

```java
double small = 10; 
double divRes = 100.0/15.0;
double big = Math.pow(2, 1023);
double bigger = Math.pow(2, 1024);
System.out.println(small); // Tam sayıyı da float haline çevirir. 
System.out.println(divRes);
System.out.println(big);
System.out.println(bigger); // Maximum sayıdan sonraki sayılar için Infinity demeye başlar. 
```

Javada normal division yaparken girdiğimiz sayılar default olarak int türüne sahip olduğu için sonuç da int değerine sahip olacaktır. Ancak sayıları double türünde girersek(ya daha önce belirterek ya da sayıları float halinde yazarak) sonuç da double türünde(float halinde olacaktır).

```java
int n1 = 100/12; 
double n2 = 100/12; // double olarak belirtmemize rağmen girilen sayılar int türünde olduğu için cevap int division sonucu gibi olacaktır.  
double n3 = (double)100/(double)12;
double n4 = 100.0/12.0; //her iki sayı da double olarak girilince cevap float division gibi olacaktır. 

//int n5 = 100.0/12.0; Böyle bir kullanım hata verecektir. 
int n5 = (int)100.0/(int)12.0; //hata vermeyecektir. 

System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
System.out.println(n5);
```

The long data type is a 64-bit two's complement integer. Its value-range lies between -9,223,372,036,854,775,808(-2^63) to 9,223,372,036,854,775,807(2^63 -1)(inclusive). Its minimum value is - 9,223,372,036,854,775,808 and maximum value is 9,223,372,036,854,775,807. Its default value is 0. The long data type is used when you need a range of values more than those provided by int.

Example: long a = 100000L, long b = -200000L

```java
long a; 
long n1 = 2;
long n2 = -100000;
double n3d = Math.pow(2, 33);
long n3 = (long)Math.pow(2, 33); //Math.pow işlemi her zaman double türünde sayı geri döndürdüğü için long'a çeviriyoruz. 
long n4 = (long)Math.pow(2, 63);
long n5 = (long)Math.pow(2, 64); //Maximum değerden sonraki sayılarda maximum değeri geri döndürür.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n3d);
System.out.println(n4);
System.out.println(n5);
```
