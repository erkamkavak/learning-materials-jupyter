Byte tr 8-bit'lik tam saylar kapsayan trdr. Deeri -128 ile 127 arasnda deiir. Default deeri 0'dr. Byk arraylerde daha az hafza kullanmak iin kullanlr. integer trnden 4 kat daha kktr.


```java
byte a; 
byte n1 = 10; 
byte n2 = 127; 
byte n3 = (byte) 128; //128 maximum deeri at iin olmad iin byte trne evirmeliyiz.  
byte n4 = (byte) (n1 + n1); //iki tane byte'n toplam byte trnde deil int trnde olacaktr. Bu deeri tekrar byte'a evirmek zorundayz.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
```

Short tr 16-bit'lik tam saylar kapsayan trdr. Maximum deeri 32.767'dir. Default deeri 0'dr. Short tr daha az hafza kullanmak iin kullanlabilir. integer trnden 2 kat daha kktr. 

```java
short a; 
short n1 = 10; 
short n2 = (short) (Math.pow(2, 15) - 1);
short n3 = (short) Math.pow(2, 15); // maximum deerden sonra negative'e dnmeye balar.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
```

Float tr 32-bit'lik float halindeki saylar kapsar. Alabilecei maximum deer 2^127'dir. Bu deerden sonra infinity deerini geri dndrr. Daha az hafza kullanmak istiyorsak array'lerde double yerine float kullanmak daha mantkldr. Default deeri 0.0F'dir.

```java
float a; 
float n1 = 10; 
float n2 = (float)10.0; //nokta ile say tanmlandnda bu say double trne girer. Bu sayy float'a evirmeliyiz. 
float n3 = (float)Math.pow(2, 31);
float n4 = (float)Math.pow(2, 127);
float n5 = (float)Math.pow(2, 128);
float n6 = (float)100.0/(float)15.0;

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
System.out.println(n5);
System.out.println(n6);
```

Double tr ise 64-bit'lik float halinde olan saylar kapsar. Alabilecei maximum deer 1.797,693,134,862,315,7 E+308 (2^1023) saysdr. Bu deerden sonra Infinity deerini geri dndrr. Default deeri 0.0d'dir. 

```java
double small = 10; 
double divRes = 100.0/15.0;
double big = Math.pow(2, 1023);
double bigger = Math.pow(2, 1024);
System.out.println(small); // Tam sayy da float haline evirir. 
System.out.println(divRes);
System.out.println(big);
System.out.println(bigger); // Maximum saydan sonraki saylar iin Infinity demeye balar. 
```

Javada normal division yaparken girdiimiz saylar default olarak int trne sahip olduu iin sonu da int deerine sahip olacaktr. Ancak saylar double trnde girersek(ya daha nce belirterek ya da saylar float halinde yazarak) sonu da double trnde(float halinde olacaktr).

```java
int n1 = 100/12; 
double n2 = 100/12; // double olarak belirtmemize ramen girilen saylar int trnde olduu iin cevap int division sonucu gibi olacaktr.  
double n3 = (double)100/(double)12;
double n4 = 100.0/12.0; //her iki say da double olarak girilince cevap float division gibi olacaktr. 

//int n5 = 100.0/12.0; Byle bir kullanm hata verecektir. 
int n5 = (int)100.0/(int)12.0; //hata vermeyecektir. 

System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n4);
System.out.println(n5);
```

The long data type is a 64-bit two's complement integer. Its value-range lies between -9,223,372,036,854,775,808(-2^63) to 9,223,372,036,854,775,807(2^63 -1)(inclusive). Its minimum value is - 9,223,372,036,854,775,808 and maximum value is 9,223,372,036,854,775,807. Its default value is 0. The long data type is used when you need a range of values more than those provided by int.

Example: long a = 100000L, long b = -200000L

```java
long a; 
long n1 = 2;
long n2 = -100000;
double n3d = Math.pow(2, 33);
long n3 = (long)Math.pow(2, 33); //Math.pow ilemi her zaman double trnde say geri dndrd iin long'a eviriyoruz. 
long n4 = (long)Math.pow(2, 63);
long n5 = (long)Math.pow(2, 64); //Maximum deerden sonraki saylarda maximum deeri geri dndrr.

System.out.println(a);
System.out.println(n1);
System.out.println(n2);
System.out.println(n3);
System.out.println(n3d);
System.out.println(n4);
System.out.println(n5);
```
