    HashMap is a collection that stores the data as (key, value) pairs, and we can access the values by their keys. HashMap has no indexes and there cannot be duplicate keys in hashMap(but there can be duplicate values). If we try to add a key that already exists then it would change the value of corresponding element. 

    HashMap should be defined as <KeyType, ValueType> (HashMap<K, V>). HashMap methods are: 
    
    get(Object key) -> O(1) time
    getOrDefault(Object key, V defaultValue) -> Returns the value to which the specified key is mapped, or defaultValue if this map contains no mapping for the key.
    
    put(K key, V value) -> O(1) time
    putIfAbsent(K key, V value) -> puts and returs null if the key is absent else returns the current value. 
    remove(Object key) -> O(1) time
    remove(Object key, Object value) -> removes the entry for the specified key if only the key is mapped to the value. 
    size() -> returns size
    containsKey(Object key) -> checks if the given key is present in the hashMap. O(1) time
    containsValue() -> checks if the given value is present in the hashMap. O(n) time (I guess) 
    keySet() -> returns the keys as a set. O(n) time
    values() -> returns the values as an ArrayList. O(n) time
    entrySet() -> returns the (key, value) pairs as a set. O(n) time
    
    Other methods: 
    compute(K key, BiFunction<>), forEach(Consumer )

    HashMap works with hash functions which converts any key Object to an integer.  It's necessary to write a hashCode() method properly for better performance in hashMap. 
    
    hashCode() method -> hashCode method is used to get the hash code of an object. In hashMap hashCode is used to calculate the bucket and therefore calculate the index.(actually using the hashCode method we are calculating the index of the key element in the hashMap array). 

    index calculation of hashMap : we generate index to minimize the size of the hashMap array. Basically the following operation is performed to calculate the index. 
                index =  hashCode(key) & (n - 1)      where n is the size of the array. 
    
    

    Example: 
    HashMap<String, Integer> map = new HashMap<>();      a hashMap with size 16 is created. 
    
    map.put(new Key("vishal"), 20);
        Calculates the hashCode of key "vishal". It will be generated as 118. Then calculates the index by index method. It will be 6. Then it creates a node linked list node as: 
        {
          int hash = 118
          Key key = {"vishal"}
          Integer value = 20
          Node next = null
        }
        then places this node at index 6, if no other node is presented there.
       
    map.put(new Key("sachin"), 30);
        Does the same thing, and finds the index as 3. Creates a node as: 
        {
          int hash = 115
          Key key = {"sachin"}
          Integer value = 30
          Node next = null
        }
        and places it at index 3. 
    
    map.put(new Key("vaibhav"), 40);
        Calculates the hashCode and finds it as 118. The index will be 6. Creates the node as: 
         {
          int hash = 118
          Key key = {"vaibhav"}
          Integer value = 40
          Node next = null
        }
        There is also another node at index 6 so this is the case of a collision. At first it checks via equals method if the keys are same and in that case changes the value of with the current value. Otherwise connects this node object via linked list and both are stored at index 6. 
     
     Now the hashMap is like: 
 
 
 
     Getting elements: 
     map.get(new Key("sachin")); 
         Calculates the hashCode of "sachin" and it will be generated as 115. Then calculates the index as 3. Goes to index 3 of array and compares the nodes of the linked list by their keys. It returns value of the node when it finds.
     It does the same thing for all keys. 
     
     Time complexities of put and get methods are ->  hashCode (O(1)) + index (O(1)) + traverse the linked list( almost O(1) ) 
     so the time complexities are almost constant. 


```python
//a hashing example

int hash(Object key){
    int h; 
    if(key == null) return 0; 
    else{
        h = key.hashCode(); //calls the object's hashCode 
        return h ^ (h >>> 16); //spreads the higher bits of hash to lower. 
    }
}

System.out.println(hash(10)); 
```
