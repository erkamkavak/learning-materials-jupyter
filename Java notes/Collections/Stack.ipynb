    In stack, elements are stored and accessed in Last In First Out manner. That is, elements are added to the top of the stack and removed from the top of the stack.



    Stack extends the Vector class, it inherits all the methods Vector.

    Stack methods
    peek() -> Retrieves but does not returns the top of stack. 
    pop() -> Retrieves and removes the top of the stack. 
    push(E e) -> Pushes an item onto the top of the stack. 
    search(Object o) 

```python
Stack<Integer> stack = new Stack<>(); 

stack.push(1); 
stack.push(2); 
stack.push(10); 

System.out.println(stack); 

System.out.println(stack.get(1));

stack.pop(); 

System.out.println(stack); 
```

```python
//Stack implementation

Class myStack {
    private int[] arr; 
    private int rearInd; 
    private int capacity; 
    private int size;
    
    public myStack(int capacity){
        this.capacity = capacity; 
        arr = new int[capacity]; 
        size = 0; 
        rearInd = -1; 
    }
        
    
    public void push(int el){
        if(size >= capacity) {
            return; 
        }
        
        rearInd++; 
        arr[rearInd] = el; 
        size++; 
    }
    
    public int pop(){
        if(size == 0){
            return -1; 
        }
        
        rearInd--; 
        size--; 
        return arr[rearInd + 1]; 
    }
    
    public int get(int index){
        if(index > rearInd) {
            return -1; 
        }
        
        return arr[index]; 
    }
    
    public String toString(){
        return Arrays.toString(
            Arrays.copyOfArray(arr, 0, rearInd + 1); 
        );
    }
    
}
```
