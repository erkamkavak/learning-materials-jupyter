    Queue data türü bir diziyi önceliklerine göre sıralamayı sağlayan stack benzeri collection'dır. Queue'lar genellikle FIFO(First in first out) mantığıyla çalışır. 
    
    FIFO queue'larında yeni elemanlar queue'nın tail'ine eklenirken remove işlemleri queue'nın head kısmından yapılır. 
    


    Queue is an interface and we need an class to implement it. There are 3 classes that implements queue. 

    
    

```java
//How to use queue

// LinkedList implementation of Queue
Queue<String> animal1 = new LinkedList<>();

// Array implementation of Queue
Queue<String> animal2 = new ArrayDeque<>();

// Priority Queue implementation of Queue
Queue<String> animal3 = new PriorityQueue<>();
```

    Queue methods
    
    add(E e) -> returns true if space is available and add is successful else returns exception. 
    element() -> retrieves but does not return the head of the queue. 
    
    offer(E e) -> inserts and returns true the specified element into the queue if it's possible else returns false
    peek() -> retrieves but does not remove the head of the queue or returns null if the queue is empty. 
    poll() -> retrieves and removes the head of the queue or returns null if the queue is empty. 
    remove() -> retrieves and removes the head of the queue
    
    default queue türünde get methodu yoktur. 

```java
Queue<Integer> queue = new LinkedList<>(); 

queue.add(5); 
queue.add(10); 
queue.add(15); 

System.out.println(queue.peek());
System.out.println(queue.poll());

System.out.println(queue);
```

```java
//Queue implementation


class Queue{
    private int[] arr; 
    private int frontInd; //front index that points to the front element 
    private int rearInd;  //rear index that points to the last element
    private int capacity; 
    private int size; 
     
    
    public Queue(int capacity){
        arr = new int[capacity]; 
        frontInd = 0; 
        rearInd = -1; 
        this.capacity = capacity; 
        size = 0; 
    }
    
    public void enqueue(int el){
        if(size >= capacity){
            return; 
        }
        
        //yeni elemanı ekliyoruz.
        rearInd++;
        arr[rearInd] = el; 
        
        size++; 
    }
    
    
    public void dequeue(){
        if(size == 0) {
            return; 
        }
        
        frontInd++; 
        
        size--; 
    }
    
    
    public int get(int index) {
        if(index > rearInd){
            return -1; 
        }
        
        return arr[index + frontInd];
    }
    
    
    public String toString(){
        return Arrays.toString(
            Arrays.copyOfRange(arr, frontInd, rearInd + 1)
        ); 
    }
}


Queue q = new Queue(5);
 
q.enqueue(1);
q.enqueue(2);
q.enqueue(3);

System.out.println(q); 

System.out.println(q.get(2)); 

q.dequeue(); 

System.out.println(q); 

System.out.println(q.get(1));

System.out.println(q.get(0)); 
```
