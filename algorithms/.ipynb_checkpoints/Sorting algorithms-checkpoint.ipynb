Selection Sort Algorithm

     It sorts an array by repeatedly finding the minimum element (considering decresing order) from unsorted part and putting it at the beginning.



```python
def selectionSort(lst):
    for i in range(len(lst)):
        ind = i 
        for j in range(i + 1, len(lst)):
            if lst[j] < lst[ind]:
                ind = j
        lst[i], lst[ind] = lst[ind], lst[i]
    return lst
print(selectionSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
selectionSort(lst)
end = timer()
print(end - start)
```

Bubble Sort Algorithm

    It works by repeatedly swapping the adjacent wrong elements.
    


```python
def bubbleSort(lst):
    swap = 1
    while swap != 0: 
        swap = 0
        for x in range(len(lst) - 1): 
            if lst[x] > lst[x + 1]:
                lst[x], lst[x + 1] = lst[x + 1], lst[x]
                swap += 1
    return lst

print(bubbleSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
bubbleSort(lst)
end = timer()
print(end - start)
```

Insertion Sort Algorithm

    Values from unsorted part is picked and placed at the sorted position.



```python
#first way
def insertionSort(lst):
    for i in range(len(lst)):
        curr = lst[i]
        j = i - 1
        while j >= 0 and curr < lst[j]:
            lst[j], lst[j + 1] = lst[j + 1], lst[j]
            j -= 1
    return lst

#second way
def insertionSort2(lst):
    for i in range(len(lst)):
        curr = lst.pop(i)
        j = i - 1
        while j >= 0 and curr < lst[j]:
            j -= 1
        lst.insert(j + 1, curr)
    return lst

print(insertionSort([1, 5, 3, 2, 7, 8, 6, 4]))
print(insertionSort2([1, 5, 3, 2, 7, 8, 6, 4]))


start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
bubbleSort(lst)
end = timer()
print(end - start)
```

Merge Sort Algorithm

    Verilen listi iki parçaya bölerek bunları recursive olarak merge sort ile sort ettikten sonra bu iki parçayı sort edilmiş şekilde birleştiren algoritmadır. 
    


```python
from collections import deque
#recursive çözümü
def mergeSort(lst):
    if len(lst) == 1: 
        return deque(lst)
    pivot = len(lst)//2
    leftSide = mergeSort(lst[:pivot])
    rightSide = mergeSort(lst[pivot:])
    res = deque()
    while leftSide and rightSide: 
        if leftSide[0] < rightSide[0]:
            res.append(leftSide.popleft())
        else: 
            res.append(rightSide.popleft())
    res += leftSide
    res += rightSide
    return res
print(mergeSort([1, 5, 3, 2, 7, 8, 6, 4]))

# def mergeSort2(stack):
#     i = 1
#     previous = stack[0:i]
#     place = 1
#     while place < len(stack):
#         leftSide = previous
#         rightSide = stack[place:min(place + i, len(lst))]
#         sorted_l = []
#         while leftSide and rightSide: 
#             if leftSide[0] > rightSide[0]:
#                 sorted_l.append(leftSide.pop(0))
#             else: 
#                 sorted_l.append(rightSide.pop(0))
#         sorted_l += leftSide
#         sorted_l += rightSide
#         previous = sorted_l
#         i *= 2
#         place = place + i
#     return previous
    

print(mergeSort2([1, 5, 3, 2, 7, 8, 6, 4]))


start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
mergeSort(lst)
end = timer()
print(end - start)
```

Quick Sort Algorithm

    It is based on Divide and Conquer algorithm. It picks an element as pivot and partitions the given array around the picked pivot.



```python
def quickSort(lst):
    if len(lst) == 0:
        return []
    pivot = lst.pop()
    leftSide = []
    rightSide = []
    for x in lst:
        if x < pivot: 
            leftSide.append(x)
        else: 
            rightSide.append(x)
    res = quickSort(leftSide) + [pivot] + quickSort(rightSide)
    return res

print(quickSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
quickSort(lst)
end = timer()
print(end - start)
```

Heap Sort Algorithm

     It is a comparison based sorting technique based on Binary Heap data structure.
    
    
    Heap Sort has O(nlog n) time complexities for all the cases ( best case, average case, and worst case).

```python
import heapq 

def heapSort(lst):
    size = len(lst)
    heapq.heapify(lst)
    res = []
    for x in range(size):
        res.append(heapq.heappop(lst))
    return res

print(heapSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
heapSort(lst)
end = timer()
print(end - start)
```

Shell Sort Algorithm

    Shell sort is an algorithm that first sorts the elements far apart from each other and successively reduces the interval between the elements to be sorted. It is a generalized version of insertion sort.
    
    In shell sort, elements at a specific interval are sorted. The interval between the elements is gradually decreased based on the sequence used. The performance of the shell sort depends on the type of sequence used for a given input array.

    Some of the optimal sequences used are:

    Shell's original sequence: N/2 , N/4 , …, 1
    Knuth's increments: 1, 4, 13, …, (3k – 1) / 2
    Sedgewick's increments: 1, 8, 23, 77, 281, 1073, 4193, 16577...4j+1+ 3·2j+ 1
    ...
    
    Worst case complexity is always less than or equal to O(n2).
    Best Case Complexity: O(n*log n)
    Average Case Complexity: O(n*log n)

```python
def shellSort(array):
    size = len(array)
    # Rearrange elements at each n/2, n/4, n/8, ... intervals
    interval = size // 2
    while interval > 0:
        for i in range(interval, size):
            temp = array[i]
            j = i
            while j >= interval and array[j - interval] > temp:
                array[j] = array[j - interval]
                j -= interval

            array[j] = temp
        interval //= 2
    return array

print(shellSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
shellSort(lst)
end = timer()
print(end - start)
```

Bucket Sort Algorithm 

    Bucket Sort is a sorting technique that sorts the elements by first dividing the elements into several groups called buckets. The elements inside each bucket are sorted using any of the suitable sorting algorithms or recursively calling the same algorithm.

    Bucket sort is mostly used when there are floating point values


bucketSort()

  create N buckets each of which can hold a range of values

  for all the buckets
  
   initialize each bucket with 0 values
  
  for all the buckets
  
   put elements into buckets matching the range
  
  for all the buckets 

   sort elements in each bucket
  
  gather elements from each bucket

end bucketSort

```python
def bucketSort(array):
    bucket = []

    for i in range(len(array)):
        bucket.append([])

    for j in array:
        index_b = int(10 * j)
        bucket[index_b].append(j)

    for i in range(len(array)):
        bucket[i] = sorted(bucket[i])

    k = 0
    for i in range(len(array)):
        for j in range(len(bucket[i])):
            array[k] = bucket[i][j]
            k += 1
    return array

print(bucketSort([.42, .32, .33, .52, .37, .47, .51]))
```

Counting Sort Algorithm

    Counting sort is a sorting algorithm that sorts the elements of an array by counting the number of occurrences of each unique element in the array.
    


Find out the maximum element (let it be max) from the given array.

Initialize an array of length max+1 with all elements 0. This array is used for storing the count of the elements in the array.



```python
def countingSort(lst):
    maxE = max(lst)
    cnt = [0]*(maxE + 1)
    for x in lst:
        cnt[x] += 1
    res = []
    for x in range(len(cnt)): 
        res += [x]*cnt[x]
    return res

print(countingSort([1, 5, 3, 2, 7, 8, 6, 4]))

start = timer()
lst = [random.randrange(0,100000) for i in range(999)]
countingSort(lst)
end = timer()
print(end - start)
```

Radix Sort Algorithm

    Radix sort is a sorting technique that sorts the elements by first grouping the individual digits of the same place value. Then, sort the elements according to their increasing/decreasing order.
    Counting sort is used as an intermediate sort in radix sort. (It is upgraded form of counting sort)



```python
def countingSortForRadix(lst, place):
    #We will only need an array of size 10 for storing counts.
    size = len(lst)
    cnts = [0]*10
    
    # Calculate count of elements
    for x in lst: 
        #we can find the number in the place(or index in counts array) by number//place % 10
        #ex:  number = 105 - place = 1  ->  105//1 % 10 = 5
        #     number = 105 - place = 100  -> 105//100 % 10 = 1
        index = x//place % 10  
        cnts[index] += 1
    
   # Calculate cummulative count
    for i in range(1, 10):
        cnts[i] += cnts[i - 1]

    # Place the elements in sorted order
    output = [0]*size
    i = size - 1
    while i >= 0:
        index = lst[i] // place % 10
        output[cnts[index] - 1] = lst[i]
        cnts[index % 10] -= 1
        i -= 1

    for i in range(0, size):
        lst[i] = output[i]

def radixSort(lst):
    #we should first find the number of iterations we should do based on the maximum value in the array.
    maxPlace = len(str(max(lst))) #length of this number will give us the iteration number.
    
    place = 1 
    for _ in range(maxPlace):
        countingSortForRadix(lst, place)
        place *= 10
    
    return lst

print(radixSort([121, 432, 564, 23, 1, 45, 788]))


def radixSort2(nums):
    maxD = len(str(max(nums)))
    size = len(nums)

    place = 1
    for x in range(maxD):
        cnts = [0]*10

        for x in nums: 
            index = x//place % 10
            cnts[index] += 1
            
        for x in range(1, len(cnts)): 
            cnts[x] += cnts[x - 1]

        res = [0]*size
        for x in nums[::-1]:
            index = x // place % 10 
            res[cnts[index] - 1] = x
            cnts[index] -= 1
            
        for i in range(size):
            nums[i] = res[i]

        place *= 10
    return nums

print(radixSort2([121, 432, 564, 23, 1, 45, 788]))
```
