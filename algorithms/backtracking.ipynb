Knight Tour problem

    Bir satranç tahtasında(N*N) ilk kareye(1, 1) konulan knight'ın tüm satranç tahtasını dolaşmasını sağlayan(her kareyi bir kere geçebilir) yol haritasını bulun ve her kareyi geçme sırasını yazdırın. 
   

```python
N = 8

def isSafe(x, y, board):
    #satranç tahtasındaki x, y noktasının boş olup olmadığını kontrol ediyoruz. 
    if 0 < x < N and 0 < y < N and board[x][y] == -1: 
        return True
    return False

def printSolution(n, board):
    #satranç tahtasını print ediyoruz. 
    for x in range(n):
        for y in range(n):
            print(board[x][y], end = " ")
        print()
        
def solveKT(N):
    #ilk önce board'ı oluşturuyoruz. 
    board = [[-1 for _ in range(N)] for _ in range(N)]
    
    board[0][0] = 1
    
    #knight'ın yapabileceği hareketleri yazıyoruz. knight 8 farklı yere hareket edebilir.
    move_x = [2, 1, -1, -2, -2, -1, 1, 2]
    move_y = [1, 2, 2, 1, -1, -2, -2, -1]
    
    #knight'ın kaçıncı harekette olduğunu pos ile tutacağız.
    pos = 1
    
    if not solveKT_rec(board, 0, 0, move_x, move_y, pos):
        print("Sorunun çözümü bulunamıyor.")
    else: 
        printSolution(N, board)

        
def solveKT_rec(board, curr_x, curr_y, move_x, move_y, pos):
    #base condition
    if pos == N**2: 
        return True
    
    #tüm hareketleri ayrı ayrı kontrol ediyoruz.
    for i in range(N):
        #her hareket için yeni x ve y değerlerini kontrol ediyoruz.
        new_x = curr_x + move_x[i]
        new_y = curr_y + move_y[i]
        if isSafe(new_x, new_y, board):
            #eğer seçtiğimiz nokta güvenli ise bu noktadan devam ediyoruz.
            board[new_x][new_y] = pos
            if solveKT_rec(board, new_x, new_y, move_x, move_y, pos + 1):
                return True
            else: 
                #backtracking kısmı > eğer seçtiğimiz noktadan sonra bir yol bulamadıysak board'ı eski haline döndürüyoruz ve kaldığımız yerden devam ediyoruz.
                board[new_x][new_y] = -1
    #eğer hareketlerin hiçbiri için bir yol bulamadıysak False döndürüyoruz.
    return False

solveKT(N)
```

```python
from collections import defaultdict


def add_edge(graph, vertex_a, vertex_b):
    graph[vertex_a].add(vertex_b)
    graph[vertex_b].add(vertex_a)


def build_graph(board_size):
    graph = defaultdict(set)
    for row in range(board_size):
        for col in range(board_size):
            for to_row, to_col in legal_moves_from(row, col, board_size):
                add_edge(graph, (row, col), (to_row, to_col))
    return graph

MOVE_OFFSETS = (
              (-1, -2), ( 1, -2),
    (-2, -1),                     ( 2, -1),
    (-2,  1),                     ( 2,  1),
              (-1,  2), ( 1,  2),
)


def legal_moves_from(row, col, board_size):
    for row_offset, col_offset in MOVE_OFFSETS:
        move_row, move_col = row + row_offset, col + col_offset
        if 0 <= move_row < board_size and 0 <= move_col < board_size:
            yield move_row, move_col

def first_true(sequence):
    for item in sequence:
        if item:
            return item
    return None


def find_solution_for(board_size, heuristic=lambda graph: None):
    graph = build_graph(board_size)
    total_squares = board_size * board_size

    def traverse(path, current_vertex):
        if len(path) + 1 == total_squares:
            # including the current square, we've visited every square,
            # so return the path as a solution
            return path + [current_vertex]

        yet_to_visit = graph[current_vertex] - set(path)
        if not yet_to_visit:
            # no unvisited neighbors, so dead end
            return False

        # try all valid paths from here
        next_vertices = sorted(yet_to_visit, key = heuristic(graph))
        return first_true(traverse(path + [current_vertex], vertex)
                          for vertex in next_vertices)

    # try to find a solution from any square on the board
    return first_true(traverse([], starting_vertex)
                      for starting_vertex in graph)

find_solution_for(6)
```

    Given strings S_1,S_2,S_3 consisting of lowercase English letters, solve the alphametic S_1+S_2=S_3.

```python
def solveEquation(s1, s2, s3): 

    chars = []
    for c in s1: 
        if not c in chars: chars.append(c)
    for c in s2: 
        if not c in chars: chars.append(c)
    for c in s3:
        if not c in chars: chars.append(c)


    charToNum = {}

    def findNum(string): 
        s = 0
        for i in range(len(string)): 
            s = (s * 10) + charToNum[string[i]]
        return s


    def backtrack(i, nums): 
        if i == len(chars): 
            return findNum(s1) + findNum(s2) == findNum(s3)

        for j in range(len(nums)): 
            if (nums[j] == -1) or (j == 0 and (s1[0] == chars[i] or s2[0] == chars[i] or s3[0] == chars[i]) ): 
                continue

            charToNum[chars[i]] = nums[j]

            nums[j] = -1
            if backtrack(i + 1, nums): 
                return True

            nums[j] = j

        return False

    backtrack(0, [*range(10)])
    print(findNum(s1), findNum(s2), findNum(s3))
    
# send + more = money eşitliğini çözer
solveEquation("send", "more", "money")
```
